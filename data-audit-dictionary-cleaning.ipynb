```python
def clean_columns(df):

    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_", regex=False)
        .str.replace("-", "_", regex=False)
        .str.replace("/", "_", regex=False)
        .str.replace("(", "", regex=False)
        .str.replace(")", "", regex=False)
    )

    return df
```

```python
def strip_strings(df):

    for col in df.select_dtypes(include="object"):

        df[col] = df[col].str.strip()

    return df
```

```python
def remove_duplicates(df):

    return df.drop_duplicates()
```

```python
def remove_empty_rows(df):

    return df.dropna(how="all")
```

```python
def remove_empty_columns(df):

    return df.dropna(axis=1, how="all")
```

```python
COUNTRY_MAPPING = {

    "Egypt, Arab Rep.":"Egypt",

    "Arab Republic of Egypt":"Egypt",

    "United Republic of Tanzania":"Tanzania",

    "Congo, Dem. Rep.":"Democratic Republic of the Congo",

    "Congo, Rep.":"Republic of the Congo",

    "Côte d'Ivoire":"Ivory Coast"

}
```

```python
def standardize_countries(df):

    for col in df.columns:

        if "country" in col:

            df[col] = df[col].replace(COUNTRY_MAPPING)

    return df
```

```python
AFRICAN_COUNTRIES = [
    "Algeria","Angola","Benin","Botswana","Burkina Faso","Burundi",
    "Cabo Verde","Cameroon","Central African Republic","Chad","Comoros",
    "Democratic Republic of the Congo","Republic of the Congo","Djibouti",
    "Egypt","Equatorial Guinea","Eritrea","Eswatini","Ethiopia","Gabon",
    "Gambia","Ghana","Guinea","Guinea-Bissau","Ivory Coast","Kenya",
    "Lesotho","Liberia","Libya","Madagascar","Malawi","Mali","Mauritania",
    "Mauritius","Morocco","Mozambique","Namibia","Niger","Nigeria",
    "Rwanda","Sao Tome and Principe","Senegal","Seychelles",
    "Sierra Leone","Somalia","South Africa","South Sudan","Sudan",
    "Tanzania","Togo","Tunisia","Uganda","Zambia","Zimbabwe"
]
```

```python
def filter_africa(df):

    for col in df.columns:

        if "country" in col:

            return df[df[col].isin(AFRICAN_COUNTRIES)]

    return df
```

***WHO Cleaning***

```python
def clean_who(df):

    df = clean_columns(df)

    df = strip_strings(df)

    df = remove_duplicates(df)

    df = remove_empty_columns(df)

    df = remove_empty_rows(df)

    df = standardize_countries(df)

    df = filter_africa(df)

    return df
```

```python
def transform_who(df, value_name):
    """
    Cleans WHO indicator datasets into a standard long format.

    Parameters
    ----------
    df : pandas.DataFrame
        Raw WHO dataset.

    value_name : str
        Name for the numerical indicator column.

    Returns
    -------
    pandas.DataFrame
    """

    keep_columns = [
        "location",
        "spatialdimvaluecode",
        "period",
        "factvaluenumeric"
    ]

    df = df[keep_columns].copy()

    df = df.rename(columns={
        "location": "country",
        "spatialdimvaluecode": "iso3",
        "period": "year",
        "factvaluenumeric": value_name
    })

    df["country"] = df["country"].str.strip()

    df["iso3"] = df["iso3"].str.strip()

    df["year"] = pd.to_numeric(df["year"], errors="coerce").astype("Int64")

    df[value_name] = pd.to_numeric(
        df[value_name],
        errors="coerce"
    )

    df = df.drop_duplicates()

    df = df.sort_values(
        ["country", "year"]
    ).reset_index(drop=True)

    return df
```

```python
# ==========================================================
# Clean IHME GBD Dataset
# ==========================================================

def transform_gbd(df):

    keep_columns = [
        "measure_name",
        "location_name",
        "sex_name",
        "age_name",
        "cause_name",
        "metric_name",
        "year",
        "val",
        "upper",
        "lower"
    ]

    df = df[keep_columns].copy()

    df = df.rename(columns={
        "measure_name":"measure",
        "location_name":"country",
        "sex_name":"sex",
        "age_name":"age",
        "cause_name":"cause",
        "metric_name":"metric",
        "val":"value",
        "upper":"upper_ci",
        "lower":"lower_ci"
    })

    df["country"] = df["country"].str.strip()

    df["measure"] = df["measure"].str.strip()

    df["cause"] = df["cause"].str.strip()

    df["metric"] = df["metric"].str.strip()

    df["sex"] = df["sex"].str.strip()

    df["age"] = df["age"].str.strip()

    df["year"] = pd.to_numeric(
        df["year"],
        errors="coerce"
    ).astype(int)

    numeric_cols = [
        "value",
        "upper_ci",
        "lower_ci"
    ]

    for col in numeric_cols:
        df[col] = pd.to_numeric(
            df[col],
            errors="coerce"
        )

    df = df.drop_duplicates()

    df = df.sort_values(
        ["country","year"]
    ).reset_index(drop=True)

    return df
```
